```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]# 최고 기온값을 저장할 리스트
low=[] # 최저 기온값을 저장할 리스트

for row in data:
    if row[-1] !='' and row[-2] !='' :    # 최고 기온 데이터 값이 존재하면
        if 1983<=int(row[0].split('-')[0]): # 1983년 이후 데이터라면
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '16':
                high.append(float(row[-1]))   # 최고 기온값을 high에 추가
                low.append(float(row[-2]))   # 최저 기온값을 low에 추가
        
plt.figure(figsize=(10,2))   
plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend()
plt.show()

f.close()
```

```python
# 1907~2019: 서울 최고온도 빈도

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]      
aug=[]

for row in data:
    if row[-1] !='' :
        if row[0].split('-')[1]=='01':
            jan.append(float(row[-1]))   # 최고 온도만 추가
        if row[0].split('-')[1]=='08':
            aug.append(float(row[-1]))   # 최고 온도만 추가
        
        
plt.rc('font',family='Malgun Gothic') # 한글 깨짐 방지
plt.rcParams['axes.unicode_minus'] = False #-부호 깨짐 방지        
        
plt.hist(jan,bins=100,color='b',label='Jan')
plt.hist(aug,bins=100,color='r',label='Aug')
plt.legend(loc=2)
plt.title('1907-2019: 1월, 8월 최고온도')
                            
plt.show()

f.close()
```
