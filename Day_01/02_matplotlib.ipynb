```python
import matplotlib.pyplot as plt
plt.title('legned')#제목넣기
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='dsc')
plt.legend(loc = 0) #범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('color')#제목넣기
plt.plot([10,20,30,40],color='skyblue',label='skyblue')
plt.plot([40,30,20,10],color='pink',label='pink')
plt.legend(loc = 0) #범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('linestyle')#제목넣기

#red dashed 그래프
plt.plot([10,20,30,40],color='r',linestyle='--',label='dashed')
# plt.plot([10,20,30,40],color='r','r--',label='dashed')

# green dotted 그래프
plt.plot([40,30,20,10],color='b',ls=':',label='dotted')
# plt.plot([40,30,20,10],'b:',ls=':',label='dotted')

plt.legend(loc = 0) #범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.title('marker')#제목넣기

plt.plot([10,20,30,40],'r.',label='circle')# red 원형 마커

plt.plot([40,30,20,10],'b^',label='triangle up')# green 삼각형 마커

plt.legend(loc = 0) #범례표시
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

dice = []
for i in range(6):
    dice.append(random.randint(1,6)) # 1~6 사이의 난수
print(dice)
plt.hist(dice, bins=6) # bins : 가로축 구간 개수
plt.show()
```
