```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
high=[]#    
low=[] #    

for row in data:
    if row[-1] !='' and row[-2] !='' :    #     
        if 1983<=int(row[0].split('-')[0]): # 1983  
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '16':
                high.append(float(row[-1]))   #   high 
                low.append(float(row[-2]))   #   low 
        
plt.figure(figsize=(10,2))   
plt.plot(high,'hotpink',label='high')
plt.plot(low,'skyblue',label='low')
plt.legend()
plt.show()

f.close()
```

```python
# 1907~2019:   

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]      
aug=[]

for row in data:
    if row[-1] !='' :
        if row[0].split('-')[1]=='01':
            jan.append(float(row[-1]))   #   
        if row[0].split('-')[1]=='08':
            aug.append(float(row[-1]))   #   
        
        
plt.rc('font',family='Malgun Gothic') #   
plt.rcParams['axes.unicode_minus'] = False #-          
        
plt.hist(jan,bins=100,color='b',label='Jan')
plt.hist(aug,bins=100,color='r',label='Aug')
plt.legend(loc=2)
plt.title('1907-2019: 1, 8 ')
                            
plt.show()

f.close()
```
