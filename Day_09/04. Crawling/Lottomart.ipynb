```python
from bs4 import BeautifulSoup        # 크롤링 라이브러리
import urllib.request as req         # 웹에서 데이터 받아오는 라이브러리|
from datetime import datetime as dt  # 현재 시간을 알려주는 라이브러리
```

```python
url = "http://company.lottemart.com/bc/branchSearch/branchSearch.do?currentPageNo={}&schBrnchTypeCd=BC0701"
```

```python
market_name_query = "#contents > ul > li > div > div.article1 > h3"
address_query =  "#contents > ul > li > div > div.article2 > ul > li"
```

```python
market_address_new = []
market_address_old = []
market_names = []

for i in range(1,26):
    target = url.format(i)
    
    res = req.urlopen(target)

    # HTML 분석하기
    data = BeautifulSoup(res, "html.parser") # HTML 형식으로 데이터 변환
    
    market_name = data.select(market_name_query)
    address = data.select(address_query)
    
    for num, add in enumerate(address):
        
        if num % 2 == 0:
            market_address_new.append(add.text)
        else:
            market_address_old.append(add.text)
    
    for name in market_name:
        market_names.append(name.text)
```

```python
print(market_address_new, len(market_address_new))
```

```python
print(market_address_old, len(market_address_old))
```

```python
print(market_names, len(market_names))
```
