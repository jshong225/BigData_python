```python
import pandas as pd

df = pd.read_csv('./전국cctv표준데이터.csv', encoding='cp949')
df = df[['위도', '경도','소재지지번주소']]
df.head()

loc = []


lat = df['위도'].values
lon = df['경도'].values

for idx in range(df.shape[0]):
    loc.append([lat[idx], lon[idx]])
    
address = df['소재지지번주소'].values

len(loc), len(address)
```

```python
# 남한의 중심 좌표 (충청북도 괴산군 청천면 이평리 350, 36.6583136,127.8590182)
m = folium.Map(
  location=[36.6583136,127.8590182],
  zoom_start=7
)

from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(m)


for i in range(len(loc)):
    folium.Marker(
        location=loc[i],
        popup=address[i],
        icon=folium.Icon(color='blue'),
    ).add_to(marker_cluster)
```

```python
import folium

m = folium.Map(
    location=[37.5838699,127.0565831],
    zoom_start=10
)

import json
with open('./seoul_municipalities_geo.json',mode='rt',encoding='utf-8') as f:
    geo = json.loads(f.read())
    f.close()

folium.GeoJson(
    geo,
    name='seoul_municipalities'
).add_to(m)
```
