```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline

sns.set_style("whitegrid")
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
tips = sns.load_dataset("tips") # seaborn 라이브러리에서 제공하는 연습용 데이터 세트
tips.head(5)
```

```python
plt.figure()
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure()
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure()
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
plt.show()
```

```python
plt.figure()
sns.boxplot(x="time", y="total_bill", order=["Dinner", "Lunch"], data=tips)
plt.show()
```

```python
plt.figure()
sns.lmplot(x="total_bill", y="tip", data=tips, size=5)
plt.show()
```

```python
plt.figure()
sns.lmplot(x="total_bill", y="tip", hue="smoker",data=tips, size=5)
plt.show()
```

```python
plt.figure()
sns.lmplot(x="total_bill", y="tip", hue="smoker",markers=["o","x"],data=tips, size=5)
plt.show()
```

```python
plt.figure()
sns.lmplot(x="total_bill", y="tip", col="smoker",data=tips, size=5)
plt.show()
```

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
# pivot 함수를 이용하여 month를 index로 year를 column으로 하고, 승객수를 value로 변경.
flights = flights.pivot("month", "year", "passengers")
flights.head(5)
```

```python
plt.figure()
sns.heatmap(flights)
plt.show()
```

```python
plt.figure()
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```
