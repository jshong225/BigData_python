```python

import openpyxl
import nltk
from openpyxl import load_workbook
from nltk.sentiment.vader import SentimentIntensityAnalyzer #감정분석
from nltk.tokenize import word_tokenize,sent_tokenize #토큰 생성
from nltk.tag import pos_tag # 형태소 태그생성
from nltk import FreqDist #빈도수 측정
from wordcloud import WordCloud #워드 클라우드
import matplotlib.pyplot as plt

sid = SentimentIntensityAnalyzer()
```

```python
def open_workbook(file = None): 
  workbook = load_workbook(file, data_only=True)#엑셀 열기
  return workbook
```

```python
def get_comments(workbook,sheet,start=1) :
  worksheet = workbook[sheet]#엑셀의 시트
  result = []

  row_num = worksheet.max_row#시트 줄수
  col_num = worksheet.max_column#시트 열수
  if row_num<start:
    return result

  for r in range(start, row_num+1) :
    #result.append(worksheet['C'+str(r)].value)
    #print(start)
    #print(worksheet['C'+str(r)].value)
    result.append(worksheet.cell(r,2).value)

  return result
```

```python
def make_wcloud(file=None, sheet=None):
  wkbook = open_workbook(file)
  comment_list=get_comments(wkbook,sheet,1)
  sid = SentimentIntensityAnalyzer()
  
  
  # 문장리스트 생성
  sentence_list =[]
  for i in range(0,len(comment_list)) :
    sentence_list.append(sent_tokenize(comment_list[i])) 
  
  #print(sentence_list)

  # 토큰리스트 생성
  token_list = []
  for i in range(0,len(comment_list)) :
    token_list.append(word_tokenize(comment_list[i]))

  #print(token_list)

  #품사 태그 넣기(명사만 사용할 것이므로 원형 복원은 패스)
  tag_list = []
  for i in range(0,len(token_list)):
    tag_list.append(pos_tag(token_list[i]))

 # print(tag_list)

  #명사만 선택하자
  selected_list = []
  for i in tag_list:
    for j in i :
      if j[1] == 'NNP' and 'NN':
        score_list=sid.polarity_scores(j[0]) 
       # if score_list['pos']>0:              #긍/부정적인 단어들만 추출하기
       #     print(j[0], score_list['neg'])
        print(j[0], score_list)
        selected_list.append(j[0])
        #print(j[0],j[1])

        

  #빈도수 측정
  result = FreqDist(selected_list)

  wc = WordCloud(width=5000, height=3000, background_color="white", random_state=0)
  plt.imshow(wc.generate_from_frequencies(result))
  plt.axis("off")
  plt.show()

  return result
```
